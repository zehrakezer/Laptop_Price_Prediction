## Web scraping from Trendyol with the bs4 library

```python
from bs4 import BeautifulSoup as bts
from urllib.request import urlopen
import requests
import pandas as pd
import re
import numpy as np
```

```python
def getAndParseURL(url):
    result = requests.get(url,headers={'user-agent':"Mozilla/5.0"})
    soup = bts(result.content, 'html.parser')
    return soup
```

```python

LINK_LIST = []
for link in range(1,50):
    sayfa=getAndParseURL("https://www.trendyol.com/sr?wb=101470%2C102323%2C101606%2C101849%2C104964&wc=103108&sst=SCORE&pi="+str(link))
    for lınk in sayfa.find_all("div",{"class":"p-card-chldrn-cntnr card-border"}):
      LINK_LIST.append("https://www.trendyol.com/"+lınk.a["href"])
```

```python
LINK_LIST
```

```python
TITLE=[]
PRICE=[]
MARKA=[]
ISLEMCI=[]
SSD=[]
SISTEM=[]
EKRAN=[]
RAM=[]
ÇÖZÜNÜRLÜK=[]
EKRANHIZ=[]
EKRANKARTI=[]
EKRANHAFIZA=[]
İŞLEMCİMODEL=[]
ÇEKİRDEKSAYISI=[]
KAPASİTE=[]
HARDDİSK=[]
BOYUT=[]
EKRANBELLEKTİPİ=[]
for products in LINK_LIST[::]:
    soup = getAndParseURL(products)
    
    #GET TITLE
    try:
      for title in soup.find("h1",{"class":"pr-new-br"}):
        TITLE.append(title.text)
    except:
      TITLE.append(0)

    #GET MARKA
    try:
        MARKA.append(soup.find("h1",{"class":"pr-new-br"}).a.text)
    except:
        MARKA.append(0)

    #GET PRICE
    try:
        PRICE.append(soup.find("div",{"class":"pr-bx-nm with-org-prc"}).text.split(",")[0].replace(".","").split(" ")[0])
    except:
        PRICE.append(0)
    
    
    #GET İŞLEMCİ
    try:
        ISLEMCI.append(soup.find("div",{"class":"detail-border"}).find(text=re.compile('İşlemci Tipi')).find_next().text)
    except:
        ISLEMCI.append(0)
    
    #GET SSD
    try:
        SSD.append(soup.find("div",{"class":"detail-border"}).find(text=re.compile('SSD Kapasitesi')).find_next().text.split(" ")[0])
    except:
        SSD.append(0)
        
    #GET SISTEM
    try:
        SISTEM.append(soup.find("div",{"class":"detail-border"}).find(text=re.compile('İşletim Sistemi')).find_next().text)
    except:
        SISTEM.append(0)
        
    #EKRAN
    try:
        EKRAN.append(soup.find("div",{"class":"detail-border"}).find(text=re.compile('Ekran Kartı')).find_next().text)
    except:
        EKRAN.append(0)
        
    #RAM
    try:
        RAM.append(soup.find("div",{"class":"detail-border"}).find(text=re.compile('Ram ')).find_next().text.split(" ")[0])
    except:
        RAM.append(0)
    
    #ÇÖZÜNÜRLÜK
    try:
        ÇÖZÜNÜRLÜK.append(soup.find("div",{"class":"detail-border"}).find(text=re.compile('Çözünürlük')).find_next().text.split("x")[0])
    except:
        ÇÖZÜNÜRLÜK.append(0)
        
    #EKRAN YENİLEME HIZI
    try:
        EKRANHIZ.append(soup.find("div",{"class":"detail-border"}).find(text=re.compile('Ekran Yenileme Hızı')).find_next().text.split(" ")[0])
    except:
        EKRANHIZ.append(0)
        
    #EKRANKARTI
    try:
        EKRANKARTI.append(soup.find("div",{"class":"detail-border"}).find(text=re.compile('Ekran Kartı Tipi')).find_next().text)
    except:
        EKRANKARTI.append(0)
        
    #EKRAN HAFIZA
    try:
        EKRANHAFIZA.append(soup.find("div",{"class":"detail-border"}).find(text=re.compile('Ekran Kartı Hafızası')).find_next().text.split(" ")[0])
    except:
        EKRANHAFIZA.append(0)
        
    #İŞLEMCİ MODEL
    try:
        İŞLEMCİMODEL.append(soup.find("div",{"class":"detail-border"}).find(text=re.compile('İşlemci Modeli')).find_next().text)
    except:
        İŞLEMCİMODEL.append(0)

    #İŞLEMCİ ÇEKİRDEK SAYISI
    try:
        ÇEKİRDEKSAYISI.append(soup.find("div",{"class":"detail-border"}).find(text=re.compile('İşlemci Çekirdek Sayısı')).find_next().text)
    except:
        ÇEKİRDEKSAYISI.append(0)
        
    #KAPASİTE
    try:
        KAPASİTE.append(soup.find("div",{"class":"detail-border"}).find(text=re.compile('Kapasite')).find_next().text.split(" ")[0])
    except:
        KAPASİTE.append(0)
        
    #HARDDİSK KAPASİTE
    try:
        HARDDİSK.append(soup.find("div",{"class":"detail-border"}).find(text=re.compile('Hard Disk Kapasitesi')).find_next().text.split(" ")[0])
    except:
        HARDDİSK.append(0)
        
    #EKRAN BOYUTU
    try:
        BOYUT.append(soup.find("div",{"class":"detail-border"}).find(text=re.compile('Ekran Boyutu')).find_next().text.split(" ")[0].replace(",","."))
    except:
        BOYUT.append(0)
        
    #EKRAN KARTI BELLEK TİPİ
    try:
        EKRANBELLEKTİPİ.append(soup.find("div",{"class":"detail-border"}).find(text=re.compile('Ekran Kartı Bellek Tipi')).find_next().text)
    except:
        EKRANBELLEKTİPİ.append(0)
        
```

```python
df=pd.DataFrame(list(zip(TITLE,MARKA,PRICE,ISLEMCI,SISTEM,SSD,EKRAN,RAM,ÇÖZÜNÜRLÜK,EKRANHIZ,EKRANKARTI,EKRANHAFIZA,İŞLEMCİMODEL,ÇEKİRDEKSAYISI,KAPASİTE,HARDDİSK,BOYUT,EKRANBELLEKTİPİ)),columns=["TITLE","MARKA","PRICE","ISLEMCI","SISTEM","SSD","EKRAN","RAM","ÇÖZÜNÜRLÜK","EKRANHIZ","EKRANKARTI","EKRANHAFIZA","İŞLEMCİMODEL","ÇEKİRDEKSAYISI","KAPASİTE","HARDDİSK","BOYUT","EKRANBELLEKTİPİ"])

df
```

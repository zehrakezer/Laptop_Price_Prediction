```python
###Sklearn Kullanarak Modelleme
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression  



X1 = df.drop(columns=['PRICE_Log'])

y1 = df['PRICE_Log']
```

Our R2 value increased by 0.32

Now let's split our test into train/test/validation and find the right method.

I separated the data set as train, test and validation set. 60% of the data was reserved for train, 20% for testing and 20% for validation.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.preprocessing import StandardScaler, PolynomialFeatures

X1 = df.drop(columns=['PRICE_Log'])

y1 = df['PRICE_Log']


# Veri setinin %20'lik kısmını Test aşaması için saklıyoruz
X, X_test, y, y_test = train_test_split(X1, y1, test_size=0.2, random_state=42)
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# Creating the 3 models that we will use for the validation phase (model selection)

# Lineer Regresyon
lm = LinearRegression()

# Ridge Regresyon
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train.values)
X_val_scaled = scaler.transform(X_val.values)
X_test_scaled = scaler.transform(X_test.values)

lm_reg = Ridge(alpha=1)

# Polinomal Regresyon
poly = PolynomialFeatures(degree=2) 

X_train_poly = poly.fit_transform(X_train.values)
X_val_poly = poly.transform(X_val.values)
X_test_poly = poly.transform(X_test.values)

lm_poly = LinearRegression()
```

```python
print('X Train:', X_train.shape)
print('X Validation:', X_val.shape)
print('X test:', X_test.shape)
```

The result of regression with polynomial is very bad, we can go from ridge or linear.

```python
# Validation results

lm.fit(X_train, y_train)
print(f'Linear Regression val R^2: {lm.score(X_val, y_val):.3f}')

lm_reg.fit(X_train_scaled, y_train)
print(f'Ridge Regression val R^2: {lm_reg.score(X_val_scaled, y_val):.3f}')

lm_poly.fit(X_train_poly, y_train)
print(f'Degree 2 polynomial regression val R^2: {lm_poly.score(X_val_poly, y_val):.3f}')
```

After trying 3 different models in our train set, I chose a model based on the R-squared results. Although linear and ridge give the same result, I decided to continue with linear and train my model in that direction.

```python
lreg = LinearRegression()

lreg.fit(X_train,y_train)

pred = lreg.predict(X_train)

mse = np.mean((pred - y_train)**2)

# Hata Metrikleri
print("Train Score: ", lreg.score(X_train, y_train))
print("MSE: ", mse)
```

Train R-squared score is 0.8. This is a relatively high value. This shows that the data of our model fit well and the independent variables predict the dependent variables well.

0.06 is a relatively low MSE value, indicating that our model made accurate predictions and fitted well with the data.

```python
lreg = LinearRegression()

lreg.fit(X_train,y_train)

pred = lreg.predict(X_val)

mse = np.mean((pred - y_val)**2)

print("Validation Score: ", lreg.score(X_val, y_val))
print("MSE: ", mse)
```

There is a difference of 0.7 between the Train and Validation datasets
